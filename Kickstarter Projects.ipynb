Luckily the data set looks clean and we have very few missing values. Now I want to determine which columns precisely have missing values and what can be done about them

```python
data_18.isnull().sum(axis = 0)
```

We can see that only the name and usd pledged have missing values, and actually for the purpose of my analysis, the name column would not be used in analysis so it's missing values won't affect my analysis.
Let's study the usd pledged column and see if we could fill up the missing value or we have to drop those columns from the dataframe entirely.

```python
data_18.columns
```

Observing thre data set closely, we can see that there is a relationship between usd pledged and usd_pledged_real, usd pledged is either equal to usd_pledged real or less than. This actaully makes sense.
For the purpose of this analysis I'll assign all the values of usd_pledged_real to corresponding usd pledged that are missing.

```python
data_18.loc[data_18['usd pledged'].isnull(), 'usd pledged'] = data_18.usd_pledged_real
data_18.info()
```

Clearly the code above worked and NaN values have been filled. Now that we have eliminated NaNs we can jump right in to analysing the data set


Distributing the data set across state

```python
percentage = np.round(data_18['state'].value_counts() / len(data_18['state']) * 100, decimals=2)
percentage
```

Making the ID column the index since it won't be used in analysis but will be used to identify projects since it has unique values for each row

```python
data_18.set_index('ID', inplace =  True)
```

```python
data_18.corr()
```

The correlation above gives us a feel of the relationship between different variables, but it's not enough as the state column which is going to be my depenedent varaible for this analysis doesn't appear. Next I'll operate on the state column so I can get correlation of the variables to the state column

```python
data_18['state'].value_counts()
```

It's more reasonable to use only failed and successful in this analysis as they are clearly defined and luckily we have for analysis. So next I'll filter out my dataset leaving only successful and failed rows and then convert their values to 0 & 1, keeping successful as 1 as you would expect.

```python
completed_projects = data_18[(data_18['state'] == 'successful') | (data_18['state'] == 'failed')]
print(completed_projects.shape)

#converting state column to 1 and 0
completed_projects['state'] = (completed_projects['state'] == 'successful').astype(int)
completed_projects.head()
```

Now I'll use correlation to see the relationship between other variables and state column

```python
completed_projects.corr()
```

Great, we can see that state has an inverse relationship with goal and usd_goal_real. This makes sense as increasing the goal makes it harder to meet up with the dealine.

Now the next part is to get duration of the projects  and month, quarter and year of launch as this might affect the success of the project too

```python
#Converting date from strings into date time format
completed_projects['launched_date'] = pd.to_datetime(completed_projects['launched'], format='%Y-%m-%d %H:%M:%S')
completed_projects['deadline_date'] = pd.to_datetime(completed_projects['deadline'], format='%Y-%m-%d %H:%M:%S')

# Calculating duration from both dates
completed_projects['duration'] = (completed_projects['deadline_date'] - completed_projects['launched_date']).dt.days
#completed_projects.head()
```

```python
#creating launched quarter, month, and year
completed_projects['launched_month'] = completed_projects['launched_date'].dt.month
completed_projects['launched_quarter'] = completed_projects['launched_date'].dt.quarter
completed_projects['launched_year'] = completed_projects['launched_date'].dt.year
```

```python
completed_projects.head()
```

selecting only required columns

```python
columns = ['category','main_category','currency','goal','pledged', 'state', 'backers', 'country', 'usd pledged', 'usd_pledged_real', 'usd_goal_real','duration','launched_month','launched_quarter','launched_year']
kickstarters = completed_projects[columns]
kickstarters.head()
```

Visualizing some columns against the state cooumn to get better insights

```python
sns.countplot(x='launched_quarter', hue= 'state', data=kickstarters)
```

From this graph, roughly same number of businesses were succesful in the four quarters of the year. Therefore it can be concluded that quarter of the year does not affect the possibility of success of a project. Below I'll take a closer look at the the time of launch by looking at the month of launching each project

```python
sns.countplot(x='launched_month',hue='state',data=kickstarters)
```

Still this is inconclusive as well, as percentage of businesses that succeeded over the 12 months are not very different. At first it looks like December isn't a good time to launch a project but taking a closer look you realize that fewer projects were in fact launched in that month, so we can't come to that conclusion

```python
sns.countplot(x='launched_year',hue='state',data=kickstarters)
```

In 2009 and 2010 we had almost equal number of projects that passed and failed, however in 2015 more projects were launched and still more projects didn't succeed. What this tells us is the rate of success in 2015 is not as high, this could be due to economic reasons , anyway this could lead to further research which we could use to determine what could lead to success or failure of a start up

Now checking the correlation of the variables, it shows that increase in goal will reduce the percentage of success of a project. Let's visualize that to make it clearer

```python
sns.catplot(x='country', y='usd_goal_real', hue='state',kind='bar', data=kickstarters,height=6, aspect=2)
```

This is very clear, regardless of the country, increase in goal reduces chance of success of the project, thus the reason for the negative correlation between goal and state we saw earlier.

The remaining variables from the correlation table have a linear realationship with the state of the project.
